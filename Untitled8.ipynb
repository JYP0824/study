```python
num_samples = 10000
epochs = 20
batch_size = 32
latent_dim = 128
embed_dim = 96
```

```python
input_texts = []
target_texts = []
path = '/content/fra.txt'

lines = open('/content/fra.txt').read().split('\n')
for line in lines[:min(len(lines)-1, num_samples)]:
  input_text, target_text = line.split('\t')
  input_texts.append(input_text)

  target_text = '\t' + target_text + '\n'
  target_texts.append(target_text)
```

```python
normalized_input = []
normalized_target = []

for text in input_texts:
  tokens = re.sub(r"[^a-z0-9' ]+", " ", text.lower())
  normalized_input.append(tokens)

token_input = [word_tokenize(sentence) for sentence in normalized_input]

for text in target_texts:
  tokens = re.sub(r"[^a-z0-9' ]+", " ", text.lower())
  normalized_target.append(tokens)

token_target = [word_tokenize(sentence) for sentence in normalized_target]
```

```python
input_set = set()
target_set = set()

for txt in token_input:
  for token in txt:
    if token not in input_set:
      input_set.add(token)

for txt in token_target:
  for token in txt:
    if token not in target_set:
      target_set.add(token)
```

```python
max_encoder_seq = max(len(txt) for txt in token_input)
max_decoder_seq = max(len(txt) for txt in token_target)

print(max_encoder_seq)
print(max_decoder_seq)
```

```python
encoder_token_index = {i : char for char, i in enumerate(input_set)}
decoder_token_index = {i : char for char, i in enumerate(target_set)}
```

```python
#정수 인덱싱
encoder_index = np.zeros((10000, max_encoder_seq), dtype=np.float32)
decoder_index = np.zeros((10000, max_decoder_seq + 1), dtype=np.float32)

counter1 = 0
for txt in token_input:
    sent = np.zeros( 5 , dtype=np.float32)
    counter2 = 0
    for token in txt:
        char = encoder_token_index[token]
        sent[counter2] = char
        counter2 += 1
    encoder_index[counter1] = sent
    counter1 += 1

counter1 = 0
for txt in token_input:
    sent = np.zeros( max_decoder_seq+1 , dtype=np.float32)
    counter2 = 0
    for token in txt:
        char = encoder_token_index[token]
        sent[counter2] = char
        counter2 += 1
    decoder_index[counter1] = sent
    counter1 += 1
```

```python
from tensorflow.keras.layers import Embedding, Input, LSTM, Dense
from tensorflow.keras.models import Model

encoder_inputs = Input(shape=(max_encoder_seq,), name='encoder_input')
en_emb = Embedding(len(input_set), embed_dim, input_length=max_encoder_seq - 1)(encoder_inputs)
encoder_outputs, state_h, state_c = LSTM(latent_dim, return_state = True)(en_emb)
encoder_state = [state_h, state_c]

decoder_inputs = Input(shape=(max_decoder_seq+1,), name='decoder_input')
de_emb = Embedding(len(target_set), embed_dim, input_length=max_decoder_seq - 1)(decoder_inputs)
decoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(de_emb, initial_state=encoder_state)
decoder_dense = Dense(len(target_set), activation='softmax', name='decoder_dense')
decoder_outputs = decoder_dense(decoder_outputs)

model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(
    [encoder_index, decoder_index],  # 디코더 입력은 <start> 토큰을 제외하고 전달
    one_hot_target,  # 디코더 타겟
    batch_size=batch_size,
    epochs=epochs,
    validation_split=0.2
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,7))
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.legend()
plt.show()
```

```python
def translator(input_seq):
    states_value = encoder_model.predict(input_seq)

    target_seq = np.zeros((1, 1, len(target_set)))
    target_seq[0, 0, target_set['\t']] = 1.

    stop_condition = False
    decoded_sentence = ''

    while not stop_condition:
        output_tokens, h, c = decoder_model.predict(
            [target_seq] + states_value)


        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_char = reverse_target_char_index[sampled_token_index]
        decoded_sentence += sampled_char

        if (sampled_char == '\n' or
           len(decoded_sentence) > max_decoder_seq):
            stop_condition = True

        target_seq = np.zeros((1, 1, len(target_set)))
        target_seq[0, 0, sampled_token_index] = 1.

        states_value = [h, c]

    return decoded_sentence
```
