# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False #한글폰트 사용 시 마이너스가 깨지는 현상 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x, y, linewidth = 5)
```

```python
plt.plot(x, y, marker = 'o')
```

## 마커

```python
plt.plot(x, y, marker='o', linestyle = 'None')
```

```python
plt.plot(x, y, marker='v', markersize = 10)
```

```python
plt.plot(x, y, marker='X')
```

```python
plt.plot(x, y, marker = 'o', markersize = 20, markeredgecolor = 'red')
```

```python
plt.plot(x, y, marker = 'o', markersize = 20, markeredgecolor = 'red', markerfacecolor = 'yellow')
```

## 선 스타일

```python
plt.plot(x, y, linestyle = ':')
```

```python
plt.plot(x, y, linestyle = '--')
```

```python
plt.plot(x, y, linestyle = '-.', linewidth = 10)
```

```python
plt.plot(x, y, color = 'pink') #matplotlib colors 구글 검색
```

## 포멧

```python
plt.plot(x, y, 'ro--') #color, marker, linestyle
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x, y, 'go')
```

## 축약어

```python
plt.plot(x, y ,marker = 'o', mfc = 'red', ms = 10, mec = 'blue', ls = ':')
```

matplotlib markerfacecolor mfc하면 축약어 나옴

## 투명도

```python
 plt.plot(x, y, alpha = 0.2) #alpha : 투명도 (0~1)
```

## 그래프 크기

```python
plt.figure(figsize=(10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,5), dpi=200)
plt.plot(x, y)
```

```python
plt.figure(figsize=(10, 5), dpi = 50) #dots per inch 확대
plt.plot(x, y)
```

## 배경색

```python
plt.figure(facecolor = 'yellow')
plt.plot(x, y)
```

```python
plt.figure(facecolor='#829ded')
plt.plot(x, y)
```
