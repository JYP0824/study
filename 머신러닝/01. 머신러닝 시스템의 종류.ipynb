머신러닝은 데이터로부터 학습할 수 있는 시스템을 만드는 것으로, 학습이란 어떤 작업에서 주어진 성능 지표가 더 나아지는 것을 의미한다

### 지도 학습

훈련 데이터에 label이라는 원하는 답을 포함

classification이 전형적 지도 학습 작업이며 스팸필터가 대표적 예시이다. 스팸필터는 많은 메일 샘플과 소속 정보로 훈련되어 새 메일을 어떻게 분류할지 학습한다

또 다른 전형적 작업은 예측 변수라 불리는 feature를 사용해 target의 수치를 예측하는 것이다. 이러한 종류의 작업을 regression이라 한다. 시스템을 훈련하려면 예측 변수와 레이블이 포함된 데이터가 많이 필요하다

### 비지도 학습

비지도 학습의 훈련 데이터에는 label이 없어 시스템이 아무런 도움 없이 학습해야 한다

**군집**

예를 들어 블로그 방문자에 대한 데이터가 있다. 방문자를 비슷한 그룹으로 묶는 군집 알고리즘을 적용하면, 알고리즘이 방문자 사이의 연결고리를 찾아 만화책을 좋아하는 방문자 40%, SF를 좋아하는 방문자 20%로 나누계 된다.

**시각화**

시각화 알고리즘을 이용하면 레이블이 없는 대규모 고차원 데이터를 도식화가 가능한 2D/3D로 표현한다. 이를 통해 데이터가 어떻게 조직되어 있는지 이해할 수 있고, 새로운 패턴을 발견할 수 있다.

**차원축소**

너무 많은 정보를 잃지 않면서 데이터를 간소화하는 차원축소의 경우 상관관계가 있는 여러 특성을 하나로 합치게 된다. 예를 들어 차의 주행거리는 연식과 강하게 연관되어 있기에 차원축소 알고리즘으로 두 특성을 차의 마모 정도를 나타내는 특성으로 합칠 수 있다. 이를 특성추출이라 한다

**이상치 탐지**

학습 알고리즘에 주입하기 전에 데이터셋에서 이상한 값을 자동으로 제거하느느 방식으로, 시스템은 훈련하는 동안 정상 샘플을 만나 이를 인식하도록 훈련된다. 그 다음 새로운 샘플을 보고 정상 데이터인지 이상치인지 판단한다

**특이치 탐지**

훈련세트에  있는 모든 샘플과 달라 보이는 새로운 샘플을 탐지하는 것이 목적으로 알고리즘으로 감지하고 싶은 모든 샘플을 제거한 깨끗한 훈련 세트가 필요하다. 예를 들어 강아지 사진 수천 장이 있고, 그 중 1%가 치와와 사진이라면 특이치 탐지 알고리즘이 치와와 사진을 새로운 특이한 것으로 처리하지 못한다. 반면에 이상치 탐지 알고리즘은 이 강아지 사진을 매우 드물고 다른 강아지와 다르다 인식해 이상치로 분류하게 된다

**연관 규칙 학습**

대량의 데이터에서 흥미로운 관계를 찾는 방법으로, 편의점을 운영한다 가정하면 판매 기록에 연관 규칙을 적용하게 될 경우 라면과 핫바를 구매한 사람이 김치도 구매하는 경향이 있다는 것을 찾을 수 있게 된다.

### 준지도 학습

데이터에 레이블을 다는 것은 일반적으로 시간과 비용이 많이 들기 때문에 레이블이 없는 샘플이 많고, 레이블된 샘플은 적은 경우가 많다. 어떤 알고리즘은 일부만 레이블이 있는 데이터를 다루는데, 이를 준지도 학습이라 한다. 

구글 포토 호스팅 서비스의 경우 가족사진을 모두 올리면 사람 A는 사진 1, 5, 11에 있고, 사람 B는 사진 2, 5, 7에 있다고 자동으로 인식하게 된다. 이는 비지도 학습(**군집**)이다. 사람마다 레이블을 하나만 추가하면 사진에 있는 모든 사람의 이름을 알 수 있고, 편리하게 사진을 찾을 수 있다.대부분의 준지도 학습 알고리즘은 지도 학습과 비지도 학습의 조합으로 이루어져 있다.

### 강화 학습

강화 학습은 학습하는 시스템을 **에이전트**라고 부르며, 환경을 관찰해 행동을 실행하고, 이에 대한 보상을 받는다. 시간이 지나면서 가장 큰 보상을 얻기 위해 **정책**이라 부르는 최상의 전략을 학습하고, 정책은 주어진 상황에서 에이전트가 어떤 행동을 선택해야 할지 정의한다.

### 배치 학습

배치 학습에서는 시스템이 점진적으로 학습할 수 없다. 가용한 데이터를 모두 사용해 훈련시켜야 한다. 일반적으로 이 방식은 시간과 자원을 많이 소모하므로 보통 오프라인에서 수행하게 된다. 먼저 시스템을 훈련시키고, 그 다음 제품 시스템에 적용하면 더 이상의 학습 없이 실행된다. 즉, 학습한 것을 단지 적용만 하게 된다. 이를 **오프라인 학습**이라고 한다

배치 학습 시스템이 새로운 데이터에 대해 학습하려면, 전체 데이터를 사용하여 시스템의 새로운 버전을 처음부터 다시 훈련해야 한다. 그 후 이전 시스템을 중지시키고 새 시스템으로 교체한다. 이 방식은 시간이 오래 걸리기 때문에 시스템이 빠르게 변하는 데이터에 적응해야 한다면, 더 능동적 방법이 필요하다.

또한 전체 데이터셋을 훈련한다면 많은 컴퓨팅 자원이 필요하고, 대량을 데이터를 가지고 있는데 매일 처음부터 새로 훈련시키도록 시스템을 자동화한다면 크 비용이 발생할 것이다. 데이터 양이 아주 많은면 배치 학습 알고리즘을 사용하게 불가능 할 수 있다.

이런 경우 점진적으로 학습할 수 있는 알고리즘을 사용하는 편이 낫다

### 온라인 학습

데이털르 순차적으로 한 개씩 또는 **미니배치**라 불리는 작은 묶음 단위로 주입해 시스템을 훈련시킨다. 매 학습 단계가 빠르고, 비용이 적게 들어 시스템은 데이터가 도착하는 대로 즉시 학습할 수 있다.

온라인 학습은 연속적으로 데이터를 받고 빠른 변화에 스스로 적응해야 하는 시스템에 적합하며, 컴퓨팅 자원이 제한된 경우에도 좋다. 온라인 학습 시스템이 새로운 데이터 샘플을 학습하면 학습이 끝난 데이터는 더는 필요하지 않으므로 버리면 된다. 컴퓨터 한 대의 메인 메모리에 들어갈 수 없는 아주 큰 데이터셋의 경우 온라인 학습 알고리즘을 이용해 데이터의 일부를 읽어들이고 훈련 단계를 수행하는 과정을 전체 데이터가 모두 적용될 때까지 반복하게 된다. (점진적 학습)

**학습률**

변화하는 데이터에 얼마나 빠르게 적응할 것인지를 나타내는 파라미터로 학습률을 높게 하면 시스템이 데이터에 빠르게 적응하지만 예전 데이터를 빠르게 잊어버리며, 학습률이 낮으면 더 느리게 학습된다.

온라인 학습에서 가장 큰 문제는 시스템에 나쁜 데이터가 주입되었을 때 시스템 성능이 점진적으로 감소한다는 점으로 검색 엔진을 속여 검색 결과를 사우이에 노출시킬 수 있게 된다. 이러하 ㄴ겨웅 입력 데이터를 모니터링해 비정상 데이터를 잡아낼 수 있다

### 일반화

머신러닝 시스템은 어떻게 일반화 되는가에 따라 분류할 수 있다. 대부부느이 머신러닝 작업은 예측을 만드는 것으로, 이 말은 주어진 훈련 데이터로 학습하고, 훈련 데이터에서 본 적 없는 새로은 데이터에서 좋은 예측을 만들어야 한다.

### 사례 기반 학습

스팸 메일과 동일한 메일을 스팸이라 지정하는 대신 매우 유사한 메일을 구분하도록 프로그래밍 할 수 있다. 이렇게 하려면 두 메일 사이의 유사도를 측정해야 한다. 두 메일 사잉의 매우 간단한 유사도 측정 방법은 공통으로 포함한 단어의 수를 세는 것으로, 스팸 메일과 공통으로 가지고 있는 단어가 많으면 스팸으로 분류된다.

시스템이 훈련 샘플을 기억함으로써 학습하게 되며, 유사도 측정을 사용해 새로운 데이터와 학습한 샘플을 비교하는 식으로 일반화한다.

### 모델 기반 학습

샘플로부터 일반화시키는 또 다른 방법으로 샘플들의 모델을 만들어 예측에 사용하는 것이다.

모델 기반 학습은 데이터셋에 따라 사용되는 모델이 다르므로, 데이터셋에 적합한 모델을 찾는 **모델 선택 과정**이 필요하다. 모델에는 다양한 **모델 파라미터**가 조재하며, 데이터에 최적화되는 모델 파라미터를 찾아내는 것이 **훈련**이다. 학습 데이터셋을 훈련시키는 과정에서 선택된 모델의 성능이 얼마나 좋은지를 **적합도 함수**로 평가할 수 있으며, 얼마나 나쁜지에 대해 **비용 함수**로 평가할 수 있다.

모델 파라미터는 일반적으로 손실함수를 최소화시키는 방향으로 진행되며, 손실함수를 최소화시키는 파라미터를 찾는 과정을 훈련이라고 한다.

모델 파라미터는 머신러닝 모델이 학습 과정에서 찾아가는 파라미터이며, 사용자가 직접 설정해주는 학습률과 같은 파라미터는 모델 파라미터와 구분하기 위해 **하이퍼 파라미터**라고 한다

모델 기반 학습은 다음과 같이 진행된다.

1. 데이터 분석

2. 데이터에 적합한 모델 선택

3. 훈련 데이터로 모델 훈련 - 비용 함수를 최소화하는 모델 파라미터 탐색

4. 학습이 끝난 모델에 새로운 데이터를 적용해 예측 - 학습 데이터로 학습된 모델을 이용해 학습 데이터와 분리된 시험 데이터가 얼마나 잘 예측되는지 본다

모델 학습 기반은 모델에 종속되어 새로운 데이터를 추론하므로, 모델이 적합하지 않거나, 데이터가 부족한 경우 더 좋은 모델을 탐색해야 하고, 그 모델에 맞는 데이터를 더 수집해야 한다.

모델의 코드 작성 난이도는 낮지만, 그 모델이 구동되는 정확한 원리를 이해하지 못한다면, 제대로 된 결과를 도출하지 못할 가능성이 높다
