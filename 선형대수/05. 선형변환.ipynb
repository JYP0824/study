## Transformation

행렬에 의한 변환 함수



함수 구성 요소

정의역: 함수의 입력 변수로 들어갈수 있는 모든 값의 집합 (domain)

공역(co-domain): 함수의 출력값으로 될 수 있는 모든 output의 집합

Image: 함수의 상 (output)  // f(1) = 5, f(5) = 8

치역(Range): 공역 중 실제 함수로 쓰이는 요소만 뽑아둔 집합

하나의 함수 당 하나의 값을 가리켜야 한다
f(1)= 5, f(1) = 7   // 5, 7로 mapping이 됨 (함수x)

### Linear Transformation

정의역에서 뽑은 원소를 선형걸합 시 얻어진 벡터가 선형결합을 나중에 했을 때의 벡터가 일치 할 시 선형변환이라 부른다
