## linearly independent

해가 하나인가

해를 찾기 위해 두 벡터의 적절한 길이를 찾아(선형결합을 통해) 두 벡터의 최적의 길이 (점)을 찾음

해가 하나인 경우 = 점을 만족하기 위해 만들 수 있는 평행사변형이 하나다이다 (선형독립)

해가 여럿이 존재한다는 것 = 평행사변형을 여러개를 만들 수 있다 (선형의존)

###### 선형의존


벡터 3개의 span 범위가 겹쳐 평행사변형을 여럿 생성 할 수 있음 (해가 여럿)

벡터를 하나씩 추가해가며 span을 보았을 때 span이 늘어나는지

추가된 벡터가 전의 벡터의 span에 들어와 추가해도 span을 늘려주지 않는 상황인지 체크를 통해 선형 의존/독립 구분 가능

첫번째 벡터로 span을 만들고 벡터를 추가할 때 span 범위에 들어가는지 체크해 추가된 벡터가 span에 포함되지 않고, span 범위를 늘려주는 경우 **선형 독립적**이다

3차원 공간 상 4개의 벡터

하나하나 보지 않아도 항상 선형 의존적이다

1번째 벡터 선 / 2번째 벡터 면 / 3차원 벡터 공간을 채우게 되어 4번째 벡터는 3개의 벡터의 span안에 무조건 포함하게 된다



(3차원 상의 4개의 벡터)

선형 의존형 케이스



두 재료벡터의 방향성이 같을 때 해가 무수히 많아진다



homogenous linear system

이 때 solution을 0이 아닌 값을 가지는 것이 있을 경우 의존적



벡터를 추가했을 때 기존의 span에 들어오면 linearly dependent하다


즉 v1, v2를 span한 평면이나 v1~3로 만든 span이나 span 측면에서는 동일한 평면상에 존재하게 된다



벡터 v1~3을 통해 만들어진 span안에 b가 포함되어야 솔루션이 1개 이상 존재

어떤 형태의 한 솔루션을 찾았는데 그 값이 3 2 1 일 때 점 b와 일치 // 그런데 점 b에서 다시 원점으로 돌아올 path가 경우 의존성을 띔 (해가 여럿)





v3를 추가해도 span을 늘려주지 못함을 의미

2v1 + 3v2 - v3 = 0  (homogenious linear system) 다시 돌아옴으로 여러 solution을 가지게 됨



2 3 -1 / 5 5 0도 solution 중 하나가 됨

유일하게 하나의 평행사변형이 그려져 점(b)에 도달할 때 (solution이 하나) linearly independent하다
