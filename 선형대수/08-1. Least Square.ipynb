### Vector Norm

 벡터의 길이
 
 
 (3,4)라는 벡터의 길이






자기 자신의 내적에 루트를 씌우는 형태로 표현

(3,4)벡터에 산수배 (2배)할 경우






길이는 항상 양의 값이어야 하기에 상수에도 절대값을 씌움




(Normalization) 주어진 벡터의 방향은 바꾸지 않고 길이만 1로 바꿈

여기서 길이가 1인 벡터 (unit vector)





길이를 구한 후 그만큼 벡터에 나눠줌으로 단위벡터를 구할 수 있다
