선형변환: Nueral Net에서의 가장 기본적인 형태

 2차원에서 2차원으로 변환하는 neural net

 
  

픽셀 4개로 이루어진 이미지


입력단의 node로 들어오게 되어 4개의 dimension을 가지는 input layer

output layer는 3개 node로 가정 시 fully connected layer라 부르게 됨

일반적으로 bias term(검은거)이 추가되어 있는데, 1이라 세팅해 output에 모두 연결



입력벡터를 3차원 벡터로 만들어주는 선형변환 matrix(학습에 의해 최적화됨),  bias쪽에서 넘어온 상수벡터로 이루어짐

복잡히 연결된 fully connected layer를 기하학적으로 column의 combination으로 보면



(입력벡터(feature)에 따른 질병 발병 확률)
