### ONTO
공역 = 치역

ex)3차원 벡터를 2차원 벡터로 보내는 선형변환 시 2차원 벡터의 모든 집합(공역) 중 함수값의 집합(치역)의 범위가 같음




ONTO(전사)가 되기 위해서는 공역의 원소 개수보다 정의역의 원소 개수가 많아야 ONTO가 될 가능성이 있다. (최소 만족 조건)



2차원 벡터를 3차원 벡터로 보내는 선형 변환의 경우 ONTO가 될 수 없다



즉, 입력 벡터의 dimension이 출력 벡터보다 작으면 onto가 될 수 없음

### one-to-one

치역이 공역과 완전히 같을 필요는 없으나, 각 각 함수에 1대1로 값이 매칭되어야 한다

onto와는 반대로 3차원에서 2차원으로의 선형변환의 경우 one-to-one이 될 수 없다



T: y = Ax




b가 span에 포함도니 (근이 하나 이상 존재) (치역의 의미)

onetoone인가 아닌가를 알기 위해서는 3개의 벡터가 선형 독립인가 아닌가를 알아야함

여기서 2차원 공간에서 보다 더 많은 갯수의 벡터가 주어졌으므로 선형 독립임 (중복이 됨 / 5,7로 만들어지는 평행사변형이 여럿 (x로 만들어지는 solution 여럿))

그렇기에 one-to-one은 linearly independent해야함
