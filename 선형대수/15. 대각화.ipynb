### Diagonalization

주어진 matrix를 대각 행렬로 만든다



기본적으로 정사각 행렬이어야 함

정사각 행렬 상에서 
이와 같은 형태를 띄는 matrix이며, 주어진 입력 matrix를 diagonal matrix로 변환시키는 것을 diagonalization이라 부름

diagonal matrix로 만드는 과정은 주어진 input matrix의 오른쪽 왼쪽에 동일하지만 역행렬 관계인 matrix를 곱해준다

이를 통해 A라는 matrix를 D라는 새로운 형태의 diagonal matrix로 만들 수 있다

여기서 식을 만족하는 matrix V를 찾을 수 있으면 diagoniliza가 가능하고, V라는 matrix가 기본적으로 역행렬이 존재해야 하며, 정사각 행렬(matrix A와 동일한 타입)이어야 한다.

이 때의 matirx A를 diagonalizable A(대각화가 가능한 행렬)라 부름



v라는 matrix의 column


3x3 matrix의 경우 3개의 column(v1~3)이 있는 matrix이며,



matrix A가 정사각 행렬이기에 V역시 같은 차원의 column vector



주어진 column vector V / D는 diagonal matrix이기 때문에 non zero 값이 diagonal에만 존재

총 matrix의 첫번째 column vector는  이 linear combination coeffienct를 통해 만들어지는 벡터인데 람다1만 non zero이기에 첫번째 벡터에 람다1만 곱해줌




두 값이 같다는 것은 두 matrix의 column들이 똑같다는 것을 의미



 eigen vector / value가 만족해야 하는 식과 동일해짐

ex) matrix A에 해당하는 eigen value 3, 5

해당 eigen value에 해당하는 eigen vector





**VD**





여기서 V에 역함수가 존재할 경우 양 쪽에 곱해 D를 구할 수 있게 된다



해당 예시에서의 V는 직사각 행렬이기에 역함수가 없음

즉, V라는 matrix는 각각의 column들이 eigen vector들이 되면서, 정사각 행렬이어야 하며, 각각의 column들이 linearly independent한 column들로 구성되어야 역함수가 존재하게 된다


즉, diagonalizable하는가는 3x3 matrix의 경우 3개의 independent한 eigen vector를 찾을 수 있느냐와 같은 의미가 된다
