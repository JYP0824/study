## Subspace

3차원 공간상의 전체 집합 속 부분집합 subset

subspace = 부분집합이나 선형결합에 닫혀있는 상태

부분집합 내의 벡터들을 어떻게 선형결합을 해도 부분집합 내부에 포함된다



span안의 부분백터에 가중치를 더해 만든 두 벡터

두 백터 역시 span내에 포함되면 두 백터의 선형결합이 내부에 포함된다면 span은 subspace임을 의미한다
녀

subspace는 항상 재료벡터들의 span으로 표현될 수 있다

재료벡터에서 시작해 span을 구함

#### basis (기저벡터)

백터들의 subset이 subspace임을 의미하며 subspace는 span으로 표현될 수 있다

subspace의 기저벡터는 subspace의 전체를 뒤덮을 수 있는 벡터들

subspace상 벡터 하나를 고름 (벡터 하나로는 주어진 subspace전체를 뒤덮을 수 없음) / 벡터를 추가했더니 fully span됨 (두 벡터를 합해 기저벡터라 칭함)

단, fully span하지만 linearly independent해야한다 (중복을 허용하지 않음)

중복: 주어진 subspace의 한 point가 기저벡터의 선형결합으로 표현 시 선형결합의 가중치는 유일하게 한 가지로 존재하며, 여러 방법으로 표현할 수 없다

기저벡터는 unique하지 않다



**change of basis**

subspace 같고, 주어진 점도 같은데, 표현하는 기저벡터가 달라졌을 때 또 다른 가중치로 표현하게 됨

#### Dimension of Subspace (unique)

subspace의 기저벡터의 수를 Dimension이라 칭하며, basis는 바뀌더라도, 기저벡터의 수는 바뀌지 않는다.

subspace의 조건을 만족하는 column들의 span을 **column space**라 한다

linearly dependant한 column이 있는 경우 column space는 정의 상 3개 벡터의 span이나 2개만으로도 동일한 의미를 가진다



#### Rank of Matrix

column space의 dimension(basis의 수)를 rank라 부름
