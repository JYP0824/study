## Linear Combinations (선형 결합)

p개의 벡터에 상수배를 해 더해주는 형태

선형결합의 가중치는 실수만을 다루며, 0도 포함되기에 실제 주어진 벡터를 사용하지 않을 수 있음



재료벡터 3개를 이용 (벡터 방정식)

### Span

선형 결합 상 주어진 p개의 벡터를 선형결합해 만든 벡터

모든 가중치(실수)를 대입한 재료벡터로 만들 수 있는 모든 선형결합 벡터의 집합




3차원 공간에서 주어진 유한한 갯수의 벡터를 통해 만들수 있는 모든 선형결합의 3차원 상의 부분집합

b벡터가 주어진 재료 벡터의 span에 포함될 때 해가 존재하며, span 외의 공간상에 있는 경우 해가 존재하지 않는다


### Matrix multiplication (행렬 곱)


원래 이렇게 했으나


재료벡터에 가중치의 곱으로 선형 결합 형태로 행렬곱 계산


두 행렬 간 곱의 경우 결과 matrix의 column을 따로 계산

계수만 달리하고 선형결합을 하게됨

결과값은 결국 3개의 재료벡터의 span안에 포함됨



row벡터들의 선형결합



결과 matrix row를 따로 생각해 위의 식처럼 가중치 대입

### Sum of (Rank-1) Outer Product



두 개의 커다란 matrix를 구한 후 더함으로 matrix multiplication 진행

차원을 하나 -함(3차원 -> 2차원)으로 보다 적은 수로 근사치를 구해 계산
